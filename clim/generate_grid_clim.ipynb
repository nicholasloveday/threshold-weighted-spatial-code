```python
import xarray as xr

ds = xr.open_zarr(
    "gs://weatherbench2/datasets/era5/1959-2023_01_10-wb13-6h-1440x721_with_derived_variables.zarr"
)

precip = ds["total_precipitation_6hr"]
precip = precip.sel(time=slice("1990", "2020"))
precip_conus = precip.sel(latitude=slice(50, 24.25), longitude=slice(234, 295))

quantile_099 = precip_conus.quantile(0.99, dim="time")
quantile_0999 = precip_conus.quantile(0.999, dim="time")


quantile_099.to_netcdf("../data/thresholds/grid/clim_0.99.nc")
quantile_0999.to_netcdf("../data/thresholds/grid/clim_0.999.nc")
```
