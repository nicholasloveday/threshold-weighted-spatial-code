```python
import xarray as xr
import numpy as np
from scores.probability import tail_tw_crps_for_ensemble
```

```python
OBS_DATA_PATH = "../data/processed/obs/"
RESULTS_PATH = "../results/clim/"
```

```python
obs = xr.open_dataset(OBS_DATA_PATH)
obs = obs.rename({"valid(UTC)": "time"})
obs = obs.precip

clim_ens = obs.quantile(np.arange(0, 1, 0.01), dim="time")
thresholds_99 = xr.open_dataarray("../data/thresholds/point/clim_0.99.nc")
clim_twcrps_099 = tail_tw_crps_for_ensemble(
    clim_ens,
    obs,
    ensemble_member_dim="quantile",
    threshold=thresholds_99,
    method="fair",
    tail="upper",
    preserve_dims=["time", "station"],
    include_components=False,
)
clim_twcrps_099.to_netcdf(f"{RESULTS_PATH}clim_twcrps_099.nc")
```
