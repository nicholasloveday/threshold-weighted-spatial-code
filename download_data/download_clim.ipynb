```python
import xarray as xr
from dask.diagnostics import ProgressBar
```

```python
era5 = xr.open_zarr(
    "gs://weatherbench2/datasets/era5/1959-2023_01_10-wb13-6h-1440x721_with_derived_variables.zarr"
)
```

```python
era_precip = era5["total_precipitation_6hr"].sel(
    time=slice("1990-01-01", "2019-12-31"),
    latitude=slice(50.5, 23.5),
    longitude=slice(234.5, 293.5),
)
```

```python
with ProgressBar():
    era_precip.to_netcdf("era_6_h_precip.nc", engine="netcdf4", format="NETCDF4")
```
