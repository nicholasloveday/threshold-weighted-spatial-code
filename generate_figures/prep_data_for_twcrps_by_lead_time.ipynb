```python
path = GRAPH1_RESULTS_099_PATH
file_list = os.listdir(path)
filename = file_list[0]
daily_result = xr.open_dataarray(f"{path}{filename}")
```

```python
def combine_data(path, outpath):
    result_list = []
    file_list = os.listdir(path)
    for filename in file_list:
        if filename[-3:] != ".nc":
            continue
        daily_result = xr.open_dataarray(f"{path}{filename}")
        result_list.append(daily_result)
    results = xr.concat(result_list, dim="time")
    results.to_netcdf(outpath)
```

```python
combine_data(GRAPH1_RESULTS_CRPS_PATH, OUT_GRAPH1_RESULTS_CRPS_PATH)
print("done")
combine_data(GRAPH3_RESULTS_CRPS_PATH, OUT_GRAPH3_RESULTS_CRPS_PATH)
print("done")
combine_data(HRRR1_RESULTS_CRPS_PATH, OUT_HRRR1_RESULTS_CRPS_PATH)
print("done")
combine_data(HRRR9_RESULTS_CRPS_PATH, OUT_HRRR9_RESULTS_CRPS_PATH)
print("done")
combine_data(HRRR27_RESULTS_CRPS_PATH, OUT_HRRR27_RESULTS_CRPS_PATH)
print("done")


combine_data(GRAPH1_RESULTS_099_PATH, OUT_GRAPH1_RESULTS_099_PATH)
print("done")
combine_data(GRAPH3_RESULTS_099_PATH, OUT_GRAPH3_RESULTS_099_PATH)
print("done")
combine_data(HRRR1_RESULTS_099_PATH, OUT_HRRR1_RESULTS_099_PATH)
print("done")
combine_data(HRRR9_RESULTS_099_PATH, OUT_HRRR9_RESULTS_099_PATH)
print("done")
combine_data(HRRR27_RESULTS_099_PATH, OUT_HRRR27_RESULTS_099_PATH)
print("done")


combine_data(GRAPH1_RESULTS_0999_PATH, OUT_GRAPH1_RESULTS_0999_PATH)
print("done")
combine_data(GRAPH3_RESULTS_0999_PATH, OUT_GRAPH3_RESULTS_0999_PATH)
print("done")
combine_data(HRRR1_RESULTS_0999_PATH, OUT_HRRR1_RESULTS_0999_PATH)
print("done")
combine_data(HRRR9_RESULTS_0999_PATH, OUT_HRRR9_RESULTS_0999_PATH)
print("done")
combine_data(HRRR27_RESULTS_0999_PATH, OUT_HRRR27_RESULTS_0999_PATH)
print("done")
```
